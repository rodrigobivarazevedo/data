```python
import pandas as pd
import numpy as np

results = pd.read_csv('results.csv')
results.head()
```

#### Narrative Example

The cell below generates the example from the exercise narrative.

```python
data = {'date': ['2019-09-06', '2021-09-24', '2019-07-07', '2021-06-27', '2016-03-25', '2017-05-28', '2021-09-05', '2021-08-01', '2021-07-11'],
        'home_team': ['United States', 'El Salvador', 'United States', 'El Salvador', 'El Salvador', 'El Salvador', 'El Salvador', 'United States', 'El Salvador'],
        'away_team': ['Mexico', 'Guatemala', 'Mexico', 'Guatemala', 'Honduras', 'Honduras', 'Honduras', 'Mexico', 'Guatemala'],
        'win_margin': [-3, -2, -1, 0, 0, 0, 0, 1, 2]}
df = pd.DataFrame(data)

# create the groupby
max_wins = df.groupby(
    ['home_team', 'away_team']).agg(
    {'win_margin':'max'})

# clean the output
max_wins.columns = ['max_margin']
max_wins = max_wins.reset_index()

# show output
max_wins.head()
```

#### 1. Top 10 Most Home/Away Team Matchups

Let's count the number of matches in each home/away matchup.

- group `results` by `home_team` and `away_team`
- use the `count` aggregation to count the `date` column -- this will count the number of games in each group

We've added some code to clean the output and then display the top 10 most frequent matchups. What do you notice?

```python
## YOUR SOLUTION HERE ##
matchups = results.groupby(
    ['home_team', 'away_team']).agg(
    {'date':'count'})

# clean the groupby
matchups.columns = ['num_games']
matchups = matchups.reset_index()

# show output
matchups.sort_values('num_games',ascending=False).head(10)
```

#### 2. Top 10 Tournaments by Average Goals

Produce a table that gives the average goals per game in each `tournament` by `year`. Save your table to `tournament_goals`. Your final cleaned output should look like

|year|tournament|average_total_goals|
|---|---|---|
|2000|AFC Asian Cup qualification|4.333333|

We've added some code to display the top 10 tournaments in terms of average goals per game. What do you notice?

```python
## YOUR SOLUTION HERE ##
tournament_goals = results.groupby(
    ['year','tournament']).agg(
    {'total_goals':'mean'})

tournament_goals.columns = ['average_total_goals']
tournament_goals = tournament_goals.reset_index()

# show output
tournament_goals.sort_values('average_total_goals', ascending=False).head(10)
```
