#### Setup

Start by running the cell below to import `pandas` and `numpy`, and read-in the `results.csv` dataset.

```python
import pandas as pd
import numpy as np

results = pd.read_csv('results.csv')
results.head()
```

#### 1. Largest Win Margins by Home Teams

Group `results` by `home_team`, and calculate the largest `win_margin` for each home team. Save the results to `largest_margin`.

```python
## YOUR SOLUTION HERE ##
largest_margin = results.groupby('home_team').agg(
    {'win_margin':'max'})

# show results
largest_margin.sort_values('win_margin',ascending=False).head()
```

#### 2. All Goals in Each Tournament

Let's calculate the number of goals scored in each tournament.

Group `results` by `tournament` and calculat the sum of `total_goals` for each tournament. Save the result to the variable `tournament_goals`.

```python
## YOUR SOLUTION HERE ##
tournament_goals = results.groupby('tournament').agg(
    {'total_goals':'sum'})

# show output
tournament_goals
```

#### 3. Number of Matches each Year

Let's count the number of matches in each year.

Group `results` by `year` and calculate the `count` of the `date` column (this will count the total number of rows/matches!)

Save the result to `games_per_year`.

```python
## YOUR SOLUTION HERE ##
games_per_year = results.groupby('year').agg(
    {'date':'count'})

# show output
games_per_year
```
